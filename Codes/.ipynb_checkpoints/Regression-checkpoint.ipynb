```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd
import plotly.graph_objects as go
from sklearn.linear_model import LinearRegression


file_path = '../Dataset/Carbon Emission Calculation.xlsx'
sheet_name = "Carbon emission - Year - Type"       
df = pd.read_excel(file_path, sheet_name=sheet_name)

print(df.head())

sns.set_style("white")
reg_non_electric = LinearRegression()
reg_electric = LinearRegression()


X = df["Year"].values.reshape(-1, 1)
y_non_electric = df["Non-Elec"]
y_electric = df["Electric"]
reg_non_electric.fit(X, y_non_electric)
reg_electric.fit(X, y_electric)

future_years = np.arange(df["Year"].max() + 1, df["Year"].max() + 11).reshape(-1, 1)
pred_non_electric = reg_non_electric.predict(future_years)
pred_electric = reg_electric.predict(future_years)

df_future = pd.DataFrame({"Year": future_years.flatten(),
                          "Non-Elec ": pred_non_electric,
                          "Electric": pred_electric})
df_combined = pd.concat([df, df_future], ignore_index=True)

plt.figure(figsize=(8, 6))
sns.regplot(x="Year", y="Non-Elec", data=df, ci=95, scatter_kws={"color": "brown"}, line_kws={"color": "blue"})
plt.plot(df_future["Year"], pred_non_electric, color="red", linestyle="--", label="Prediction")
plt.scatter(df_future["Year"], pred_non_electric, color="red", label="Predicted Points")
plt.title("Non Elecric Vehicle Number in Years", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Non Electric Vehicle Number ")
plt.legend()
plt.tight_layout()
plt.savefig("benzinli_arac_regresyon_seaborn_beyaz.png")
plt.show()

plt.figure(figsize=(8, 6))
sns.regplot(x="Year", y="Electric", data=df, ci=95, scatter_kws={"color": "brown"}, line_kws={"color": "blue"})
plt.plot(df_future["Year"], pred_electric, color="red", linestyle="--", label="Prediction")
plt.scatter(df_future["Year"], pred_electric, color="red", label="Predicted Points")
plt.title("Elecric Vehicle Number in Years", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Electric Vehicle Number")
plt.legend()
plt.tight_layout()
plt.savefig("elektrikli_arac_regresyon_seaborn_beyaz.png")
plt.show()
```
