```python
import pandas as pd

file_path = "../Dataset/Yıllara_göre_araç _sayısı.xlsx"


df = pd.read_excel(file_path, header=1)
df
```

```python
df['Non-Elec'] = df['     Gasoline'] + df['   LPG'] + df['Diesel']
df
```

```python
columns_to_drop = ['(%)', '(%).1', '(%).2', '(%).3', '(%).4', '(%).5', '(%).6', 'Unknown (3)', 'Hybrid (2)'] 
df = df.drop(columns=columns_to_drop)
df
```

```python
df['Electric'] = pd.to_numeric(df['Electric'], errors='coerce').fillna(0)
```

```python
df.isnull().sum()
```

```python
df.rename(columns={'       Total': 'Total', '     Gasoline': 'Gasoline', '   LPG': 'LPG'}, inplace=True)
```
